```python
from ast import literal_eval
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl


platesem = np.array(literal_eval(open('20170504_amine_degraded_CsPbBr3_emission.txt', 'r').read()))
platesab = np.array(literal_eval(open('20170504_amine_degraded_CsPbBr3_absorption.txt', 'r').read()))

wavelengths_em_i4 = range(470,602,2)
wavelengths_ab_i4 = range(250,702,2)

times_h_i1 = [1.5,3,4.5,6,7.5,9,10,22,36,44,110,163]
# the number of hours after addition that the sample was measured

NP_ID_i2 = ['6nm_300uL', '8nm_150uL', '10nm_150uL','14nm_75uL','8nm_50uL', '8nm_100uL','8nm_200uL','8nm_400uL']
# the identity and amount of nanocrystal sample

OAStock_soln_i3 = [0,0,20,40,60,80,100,120,140,160,180,200]
# the amount of stock solution added to each well in microliters

OAstock_dilution = 135.0/(3880+135)
# the amount of microliters of oleylamine in the stock solution
```

```python
all_y = []
for a in range(0,8):
    array1 = platesab[0][a][9]
    yval = []
    for i in range(25,50):
        yval.append(array1[i])
    all_y.append(max(yval))
display(all_y)
```

```python
plt.figure(figsize=(8,8))
for i in range(len(platesab[7][1])):
    plt.plot(wavelengths_ab_i4,
             platesab[7][1][i],
             label=str(round(OAStock_soln_i3[i]*OAstock_dilution,3)) + 'uL OLA',
                      linewidth = 3)
plt.xlim(250,530)
plt.ylim(0,3)
plt.title("8nm_ at 22 hours")
plt.ylabel('Absorptivity')
plt.xlabel('Wavelength (nm)')
plt.legend()
plt.show()
```
